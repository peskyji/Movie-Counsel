```python
def make_category(x):
  movie_tag = ""
  year, duration = x

  # categorizing based on year
  if year < 1980:
    movie_tag+="before 80's"
  elif year>=1980 and year<1990:
    movie_tag+="80's"
  elif year>=1990 and year<2000:
    movie_tag+="90's"
  elif year>=2000 and year<2010:
    movie_tag+="00's"
  elif year>=2010 and year<2020:
    movie_tag+="10's"
  else:
    movie_tag+="20's"

# categorizing based on duration
  if duration <= 40:
    movie_tag+= ' short_length'
  elif duration>40 and duration<=90:
    movie_tag+=' normal_length'
  elif duration>90 and duration<=120:
    movie_tag+=' long_length'
  else:
    movie_tag+=' extra_long_length'

  return movie_tag

temp = final_df[['year','duration']].apply(make_category, axis=1)
final_df = pd.concat([final_df, pd.DataFrame(temp, columns=['movie_tags'])], axis=1)
final_df
```

```python
def create_final_movie_tags(genre, director, star, description, movie_tags):
  for x in genre.split(","):
    movie_tags += f" {x}"
  for x in director.split(","):
    movie_tags += f" {x}"
  for x in star.split(","):
    movie_tags += f" {x}"
  movie_tags += f" {description}"

  return movie_tags

final_df.movie_tags = final_df.apply(lambda row: create_final_movie_tags(row['genre'],row['directors_name'],row['stars_name'],row['description'],row['movie_tags']), axis=1)
final_df.head(4)
```
